```python
import glob

from datasets import Dataset, load_dataset
```

```python
def make_dataset1(path):
    parquet_files = glob.glob(f"{path}/*.parquet")
    return Dataset.from_parquet(parquet_files)
```

```python
def make_dataset2(path, streaming = True):
    return load_dataset("parquet", data_dir=path, split="train", streaming = streaming)
```
